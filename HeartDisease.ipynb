```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#I importojme te dhenat tona dhe i shfaqim ato.

df = pd.read_excel('heart disease.xlsx')
df
```

```python
# Duhet qe disa te dhena ti manipulojme ashtu qe mund te bejme llogari me to

df['Gender'] = df['Gender'].replace(['Female', 'Male'],[0,1])
df['ExerciseInducedAngina'] = df['ExerciseInducedAngina'].replace(['No', 'Yes'],[0,1])
df['HeartDisease'] = df['HeartDisease'].replace(['No', 'Yes'],[0,1])
```

```python
# I shfaqim te dhenat tona pasi ti kemi kryer manipulimet
df = df.drop(columns ="STDepressionByExercise")
df
```

```python
#I shohim permes metodes describe Statistikat kryesore te te dhenave tona.

df.describe()
```

```python
#Bejme nje korrelacion per te shikuar ndikimin e kolonave te njera tjetra.
#Per neve me rendesi eshte ndikimi i kolonave tjera te kolona HeartDisease.

df.corr()
```

```python
# Bejme nje paraqitje grafike, ndikimi i moshes ne heart disease.


sns.set_context("paper", font_scale = 0.8, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20}) 
sns.catplot(kind = 'count', data = df, x = 'Age', hue = 'HeartDisease', order = df['Age'].sort_values().unique())
```

```python
#Paraqitje grafike, ndikimi i Gjinise ne HeartDisease. 0- Femer, 1- Mashkull.

sns.set_context("paper", font_scale = 2, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20}) 
sns.catplot(kind = 'count', data = df, x = 'Gender', hue = 'HeartDisease')
```

```python
# Paraqitje Grafike, ndikimi i Gjinise dhe moshes ne HeartDisease.
# Femrat e Prekura nga semundjet e zemres jane me ne moshe se sa mashkujt.
# Nga ky grafik kuptohet se meshkujt kane tendece qe te preken nga semundjet e zemres ne moshe me te re se femrat.

sns.catplot(kind = 'bar', data = df, y = 'Age', x = 'Gender', hue = 'HeartDisease')
plt.title('Distribution of age vs sex with the target class')
plt.show()
```

```python
#Marrim te gjihte kolonat deri te kolona e fundit dhe i ruajme ne vektorin X.
X = df.iloc[:, :-1].values

#Marrim kolonen e fundit dhe e ruajme ne vektorin y.
y = df.iloc[:, -1].values

#I ndajme te dhenat tona 80% me 20%, ku 80% e te dhenave tona shkojne ne X_train per trajnim te modelit
# ndersa 20% shkojne per testim
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#Marrim tani Regresionin Logistic.
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter = 10000)
classifier.fit(X_train, y_train)


# Predicting the Test set results
y_pred = classifier.predict(X_test)

#Gjejme matricen e konfusionit
from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_pred, y_test)

y_pred_train = classifier.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Saktesia per setin e trainimit per Regresionin Logistik = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Saktesia per setin e testit per Regresionin Logistik = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))
```

```python
#Tani importojme klasen Stasmodel.

import statsmodels.api as sm
log_reg = sm.Logit(y_train,X_train).fit()
```

```python
print(log_reg.summary())
```

```python
# Bejme Krahasimin e y te parashikuar dhe y nga seti testues. (y eshte vektori qe permban kolonen semundjet e zemres)
y_pred = log_reg.predict(X_test)
np.set_printoptions(precision =2 )
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```
